```python
for i,row in df.iterrows():
    temp_duration = row['Duration']
    arr =temp_duration.split()
    result=0
    if len(arr)==2:
        hours = arr[0]
        minutes = arr[1]
        hours = hours.replace('h','')
        minutes = minutes.replace('m','')
        hours_int = int(hours)
        minutes_int = int(minutes)
        result = hours_int * 60 + minutes_int
    else:
        hours = arr[0]
        hours = hours.replace('h','')
        result=int(hours)
        result *= 60
    df.loc[i,'Duration']=result
```

```python
for i,row in test_df.iterrows():
    temp_duration = row['Duration']
    arr =temp_duration.split()
    result=0
    if len(arr)==2:
        hours = arr[0]
        minutes = arr[1]
        hours = hours.replace('h','')
        minutes = minutes.replace('m','')
        hours_int = int(hours)
        minutes_int = int(minutes)
        result = hours_int * 60 + minutes_int
    else:
        hours = arr[0]
        hours = hours.replace('h','')
        result=int(hours)
        result *= 60
    test_df.loc[i,'Duration']=result
```

```python
dep_var=['Price']
```

```python
cont_vars=['Duration','Date_CleanElapsed']
```

```python
cat_vars = ['Airline', 'Source_Code', 'Destination_Code', 'Travel_Score',
       'Dep_Flight_Timing_Buckets', 'Arr_Flight_Timing_Buckets',
       'Stops_Total', 'Additional_Info', 'Date_CleanYear',
       'Date_CleanMonth', 'Date_CleanWeek', 'Date_CleanDay',
       'Date_CleanDayofweek', 'Date_CleanDayofyear', 'Date_CleanIs_month_end',
       'Date_CleanIs_month_start', 'Date_CleanIs_quarter_end',
       'Date_CleanIs_quarter_start', 'Date_CleanIs_year_end',
       'Date_CleanIs_year_start']
```

```python
df.dtypes
```

```python
df.Price= df.Price.astype(float)
```

```python
df.dtypes
```

```python
for col in cat_vars:
    df[col] = df[col].astype('category')
    test_df[col]=test_df[col].astype('category')
```

```python
test_df.dtypes
```

```python
df.dtypes
```

```python
path=''
```

```python
valid_list=list(range(8000,len(df)))
```

```python
len(valid_list)
```

```python
data = (TabularList.from_df(df, path=path, cat_names=cat_vars, cont_names=cont_vars, procs=[],)
                .split_by_idx(valid_list)
                .label_from_df(cols=dep_var, label_cls=FloatList, log=True)
                .add_test(TabularList.from_df(test_df, path=path, cat_names=cat_vars, cont_names=cont_vars))
                .databunch())
```

```python
max_log_y = np.log(np.max(df['Price'])*1.2)
min_log_y = np.log(np.min(df['Price'])*0.8)
y_range = torch.tensor([min_log_y, max_log_y], device=defaults.device)
```

```python
max_log_y,min_log_y,np.exp(max_log_y),np.exp(min_log_y)
```

```python
data.show_batch(rows=10)
```

```python
learn = tabular_learner(data, layers=[1000,500], metrics=root_mean_squared_error,y_range = y_range)
```

```python
learn.model
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10)
```
