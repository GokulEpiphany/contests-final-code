```python
from fastai.vision import *
from fastai.callbacks import *
```

```python
torch.cuda.is_available()
```

```python
np.random.seed(42)
```

```python
tfms=[flip_lr(p=0.5),dihedral_affine() ],[]
```

```python
mixup=0.2
```

```python
loss_func = LabelSmoothingCrossEntropy()
```

```python
alpha=0.99
mom=0.9
eps=1e-6
```

```python
opt_func = partial(optim.Adam, betas=(mom,alpha), eps=eps)
```

```python
src = (ImageList.from_folder(path='train').split_by_rand_pct(0.2).label_from_folder())
```

```python
data = src.transform(tfms,size=(256)).databunch(bs=128).normalize().presize(512, scale=(0.35,1))
```

```python
learn = cnn_learner(data, models.densenet121,pretrained=False, metrics=[error_rate, accuracy, FBeta(beta=1,average='macro')],loss_func = loss_func,opt_func=opt_func,wd=1e-2)
```

```python
learn= learn.mixup(alpha=mixup)
```

```python
learn.bn_wd=False
```

```python
learn.true_wd=True
```

```python
learn.load('exp-5-round-5-3')
```

```python
data_320 = src.transform(tfms,size=(320)).databunch(bs=96).normalize().presize(640, scale=(0.35,1))
```

```python
data_320.show_batch(rows=3, figsize=(12,12))
```

```python
learn.data= data_320
```

```python
learn.data
```

```python
learn = learn.to_fp16(dynamic=True)
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
lr=1e-4
```

```python
learn.fit_one_cycle(20, lr, div_factor=10, pct_start=0.3)
```
