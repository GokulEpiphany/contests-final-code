```python
model = EfficientNet.from_name('efficientnet-b5')
```

```python
model_name = 'efficientnet-b5'
image_size = EfficientNet.get_image_size(model_name)
```

```python
image_size
```

```python
model = EfficientNet.from_pretrained(model_name)
```

```python
model
```

```python
np.random.seed(42)
```

```python
src = (ImageList.from_folder(path='train-snakes-cropped').split_by_rand_pct(0.2).label_from_folder())
```

```python
tfms=([rotate(degrees=(-90,90), p=0.8)],[])
```

```python
bs=8
```

```python
data = (src.transform(tfms, size=image_size, resize_method=ResizeMethod.SQUISH)
        .databunch(bs=bs).normalize(imagenet_stats))
```

```python
data.show_batch(3,figsize=(9,9))
```

```python
list(children(model))[-2:]
```

```python
model
```

```python
model.add_module('_fc',nn.Linear(2048, data.c))
```

```python
loss_func =LabelSmoothingCrossEntropy()
```

```python
RMSprop = partial(torch.optim.RMSprop,eps=1e-6,centered=True)
```

```python
learn = Learner(data, model, loss_func=loss_func, opt_func=RMSprop, metrics=[accuracy,FBeta(beta=1,average='macro')],bn_wd=False)
```

```python
learn
```

```python
learn.model
```

```python
learn.split([[learn.model._conv_stem, learn.model._bn0, learn.model._blocks[:4]],
             [learn.model._blocks[4:],learn.model._conv_head], 
             [learn.model._bn1,learn.model._fc]])
```

```python
learn.summary()
```

```python
lr=1e-3
```

```python
learn.fit_one_cycle(10,max_lr=slice(lr/100,lr))
```

```python
learn.save('b5-proper-round-1')
```

```python
learn.validate()
```
