```python
from fastai.vision import *
from fastai.callbacks import *
```

```python
torch.cuda.is_available()
```

```python
SZ = 64
```

```python
tfms = [rand_resize_crop(SZ)]+[ zoom_crop(scale=(0.75,2), do_rand=True)]
```

```python
tfms
```

```python
np.random.seed(42)
```

```python
src = (ImageList.from_folder(path='train-genus').split_by_rand_pct(0.2).label_from_folder())
```

```python
src
```

```python
data = src.transform(tfms,size=SZ).databunch(bs=384).normalize(imagenet_stats)
```

```python
data.c
```

```python
# from pathlib import Path
# path = Path("train-genus/")

# print(path)
# for c in data.classes:
#     print(c)
#     verify_images(path/c, delete=True)
```

```python
data.show_batch(rows=3, figsize=(12,9))
```

```python
mom = 0.9
alpha=0.99
eps=1e-6
```

```python
opt_func = partial(optim.Adam, betas=(mom,alpha), eps=eps)
loss_func = LabelSmoothingCrossEntropy()
```

```python
learn = cnn_learner(data, models.densenet169, metrics=[error_rate, accuracy, FBeta(beta=1,average='macro')],
                    opt_func=opt_func,loss_func = LabelSmoothingCrossEntropy(),pretrained=True,
                    wd=1e-2, bn_wd=False, true_wd=True)
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
lr=1e-2
```

```python
learn.fit_one_cycle(10,lr,div_factor=10,pct_start=0.3)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```python
interp.plot_top_losses(9, figsize=(15,11))
```

```python
interp.plot_confusion_matrix(figsize=(20,20), dpi=60)
```

```python
interp.most_confused(min_val=2)
```
