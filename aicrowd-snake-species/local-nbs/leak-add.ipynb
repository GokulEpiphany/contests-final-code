```python
from fastai.vision import *
from fastai.callbacks import *
src = (ImageList.from_folder(path='train').split_by_rand_pct(0.2).label_from_folder())
```

```python
data = src.transform(size=8).databunch(bs=24).normalize()
```

```python
leaked = pd.read_csv('merged.csv',low_memory=False)
```

```python
classes = pd.read_csv('class.csv',low_memory=False)
```

```python
classes.head()
```

```python
data_dict = classes.set_index('class_idx')['original_class'].to_dict()
```

```python
data_dict
```

```python
leaked.head()
```

```python
data.classes
```

```python
for i,row in leaked.iterrows():
    test_img = row['filename_y'].split('\\')[3]
    print(test_img)
    correct_class = row['filename_x'].split('\\')[1]
    print(correct_class)
    for j,new_row in df.iterrows():
        if new_row['filename'] == test_img:
            for c in range(0,45):
                df.loc[j,data_dict[int(data.classes[c].split("-")[1])]]=0
            df.loc[j,data_dict[int(correct_class.split("-")[1])]]=1
            break
```

```python
for i,row in df.iterrows():
    sum_temp=row[1:46].sum()
    low_limit = 1-1e-6
    high_limit = 1+1e-6
    print(sum_temp)
    assert sum_temp>= low_limit and sum_temp <= high_limit
```
