```python
import pandas as pd
```

```python
df = pd.read_csv('ladies-kurta.csv')
```

```python
df.head()
```

```python
df_needed = df['Price']
```

```python
len(df_needed)
```

```python
df_needed.head()
```

```python
df_distance = pd.DataFrame(columns=['distance'])
```

```python
df_new = df_needed.copy()
```

```python
count=0
for i in df_needed:
    distance = 0
    for j in df_new:
        distance = distance+ abs(i-j)
    df_distance.loc[count,'distance']=distance
    count=count+1
```

```python
average = sum/4366
```

```python
count=0
for i in df_distance['distance']:
    if ((i/4366)> (average*1.5)) or ((i/4366) < (average*0.1)) or (df_needed[count] < average * 0.1):
        print("Outlier "+str(count))
    count= count+1
```
